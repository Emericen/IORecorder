```python
import time
import pandas as pd
import pynput.mouse as M
import pynput.keyboard as K

mouse = M.Controller()
keyboard = K.Controller()
```

```python
def position(x, y, duration):
    time.sleep(duration)
    mouse.move(x, y)

csv_path = "recording_20250129_231839\mouse_events.csv"


# calculate timestamp difference between each row
df = pd.read_csv(csv_path)
df["delta_time"] = df["timestamp"].diff()


# calculate delta x and delta y
df["delta_x"] = df["x"].diff()
df["delta_y"] = df["y"].diff()

df.head(4000)
```

```python
fast_forward_factor = 1.2  # 1.2 is 120% speed, 0.8 is 80% speed


def position(x, y, duration):
    time.sleep(duration / fast_forward_factor)
    mouse.position = (x, y)

def move(d_x, d_y, duration):
    time.sleep(duration / fast_forward_factor)
    mouse.move(d_x, d_y)


initial_x, initial_y = df.iloc[0]["x"], df.iloc[0]["y"]
position(initial_x, initial_y, 0)

# start from the second row
for index, row in df.iterrows():
    if index == 0:
        continue
    # position(row["x"], row["y"], row["delta_time"])
    move(row["delta_x"], row["delta_y"], row["delta_time"])
    initial_x, initial_y = row["x"], row["y"]
```
